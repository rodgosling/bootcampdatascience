### Parte 1: Importando dados de uma URI

```python
# Importando bibliotecas

import pandas   as pd
import numpy    as np
from   datetime import datetime
import json
import requests
```

```python
# Importando dados do arquivo pessoas.json

#data = 'pessoas.json'


# Importando dados URI

uri = 'https://api-coleta-dados.herokuapp.com/api/pessoas'

response = requests.get(uri)
```

```python
# Carregando dados no df (arquivo pessoas.json)

#pessoas_df = pd.read_json (data)


# Carregando dados no df (URI)

if response.status_code == 200:
    pessoas_df = pd.read_json(response.content)

pessoas_df
```

```python
# Analisando tipo dos dados

pessoas_df.dtypes
```

```python
# Convertendo os valores NaN da coluna de idade para 0

pessoas_df['idade'].fillna(0, inplace=True)
pessoas_df
```

```python
# Convertendo a coluna idade para inteiro

pessoas_df['idade'] = pessoas_df['idade'].astype(int)
pessoas_df.dtypes
```

```python
# Criando coluna auxiliar ano

pessoas_df['ano'] = datetime.now().year - pessoas_df['idade']


# Corrigindo data_nasc com ano faltante

for i in range(len(pessoas_df)):
    if len(pessoas_df['data_nasc'].iloc[i]) < 10:
        anonascimento = pessoas_df['data_nasc'].astype(str) + '/' + pessoas_df['ano'].astype(str)
        pessoas_df['data_nasc'].iloc[i] = anonascimento.iloc[i] 
        
pessoas_df
```

```python
# Corrigindo pessoas sem idade

pessoas_df['ano'] = pessoas_df['data_nasc'].str.slice(start=-4)

for i in range(len(pessoas_df)):
    if pessoas_df['idade'].iloc[i] == 0:
        idade_certa = datetime.now().year - pessoas_df['ano'].astype(int)
        pessoas_df['idade'].iloc[i] = idade_certa.iloc[i] 
        
pessoas_df
```

```python
# Removendo a coluna auxiliar 'ano'

pessoas_df.drop(['ano'], axis=1, inplace = True)

pessoas_df
```

##### Importando dataframe para banco de dados MySQL

```python
# Importando bibliotecas necessárias

from sqlalchemy import create_engine
```

```python
# Criando ferramenta de conexão

engine = create_engine('mysql+pymysql://root:igti@localhost/dcbootcamp_m2')
```

```python
# Salvando dataframe no DB

pessoas_df.to_sql('pessoas', con = engine, if_exists = 'replace', index_label='id')
```

### Parte 2: Importando dados do MongoDB

```python
# Importando bibliotecas necessárias

import pandas            as pd
import json
import pymongo           # Instalar biblioteca conda install -c wakari pymongo
import pprint
import numpy             as np
import matplotlib.pyplot as plt
```

```python
# Conexão com MongoDB

con = pymongo.MongoClient('localhost', 27017)
```

```python
# Listar bancos de dados

con.list_database_names()
```

```python
# Selecionando o banco de dados

db = con.dcbootcamp
db.list_collection_names()
```

```python
# Selecionando a coleção

collection = db.modulo2
```

```python
# Selecionando o primeiro elemento da coleção

collection.find_one()
```

```python
# Salvando documentos em um dataframe

game_df = pd.DataFrame(list(collection.find({})))
game_df
```

#### Consultas

```python
# Qual jogo possui a maior venda global (“Global_Sales”) para a plataforma PS2?

game_df.loc[game_df.Platform == 'PS2'].sort_values(by = ['Global_Sales'],ascending = False).head(1)
```

```python
# Qual a posição no Rank global (“Rank”) está o segundo jogo mais vendido na Europa? Ordenar pelo atributo "EU_Sales".

game_df.sort_values(by = ['EU_Sales'], ascending = False).head(2)
```

```python
# Qual o nome do jogo que possui a menor venda global (“Global_Sales”) no ano de 2010?

game_df.loc[game_df.Year == 2010].sort_values(by = ['Global_Sales'], ascending = False).tail(1)
```

```python
# Qual a plataforma (“Platform”) do jogo mais vendido considerando as vendas globais ("Global_Sales ") no ano de 2008?

game_df.loc[game_df.Year == 2008].sort_values(by = ['Global_Sales'], ascending = False).Platform.head(1)
```

```python
# Quantos documentos existem na base de dados referentes à plataforma (“Platform”) “X360”?

game_df.loc[game_df.Platform == 'X360']._id.count()
```

```python
# Qual a plataforma (“Platform”) do jogo mais vendido no Japão ("JP_Sales") no ano de 2015?

game_df.loc[game_df.Year == 2015].sort_values(by = ['JP_Sales'], ascending = False).Platform.head(1)
```

```python
# Em que posição do “Rank” o jogo “The Witcher 3: Wild Hunt” está considerando a plataforma “XOne” no ano 2015?

xone2015 = game_df.loc[(game_df.Platform == 'XOne') & (game_df.Year == 2015)].reset_index()

for index, row in xone2015.iterrows():
    print (index, row['Name'] )
```

```python
# Em que posição do “Rank” o jogo “The Witcher 3: Wild Hunt” está considerando a plataforma “XOne” no ano 2015?

xone2015.loc[xone2015.Name == 'The Witcher 3: Wild Hunt']
```

```python
# Quantos documentos são retornados ao executar uma consulta em que o atributo “Global_Sales” seja maior ou igual a 2?

game_df.loc[game_df.Global_Sales >= 2]._id.count()
```

#### Análises

```python
game_df.describe()
```

```python
# Plataformas com maior número de jogos no ranking

game_df.groupby(['Platform'])._id.count().sort_values(ascending = False).head()
```

```python
# Plataformas com maior número de jogos no ranking

best = game_df.groupby(['Platform'])._id.count().sort_values(ascending = False).head()

best.plot(kind='bar', color = '#800080')

plt.xlabel(xlabel = None, rotation = 'horizontal')
plt.axhline(best.mean(), lw = 1, color='orange', ls = '--')
plt.title("Maior número de jogos no ranking", weight = 'bold', size = 15, color = '#800080')
plt.show()
```

```python
# Plataformas com menor número de jogos no ranking

game_df.groupby(['Platform'])._id.count().sort_values(ascending = False).tail()
```

```python
# Plataformas com menor número de jogos no ranking

worst = game_df.groupby(['Platform'])._id.count().sort_values(ascending = False).tail()

worst.plot(kind='bar', color='#800080');
plt.title("Menor número de jogos no ranking", weight = 'bold', size = 15, color = '#800080')
plt.show()
```
